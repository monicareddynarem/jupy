```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib as mp
import numpy as np
import random 
```

```python
#LINE GRAPH
x=[1,2,3,4]
y=[2,6,6,8]
plt.plot(x,y)
plt.show()
```

```python
#BAR GRAPH
x=[1,2,3,4]
y=[4,1,5,12]
c=['r','y','b','g']
plt.xlabel("x axis values-no.of students",fontsize=20)
plt.ylabel("yaxis values-grades",)
plt.title("students and grades")
plt.bar(x,y,color=c,width=0.4,align="edge",edgecolor='r',
        linewidth=5,linestyle=":",alpha=0.4,label="mygraph")
plt.legend()
#label need plt.legend()
#alpha=intensity of colors 0< <1
#linestyle-":" to et dotted line
#align(edge/centre) changes the position of x axis values wrto bars
#width is the width of bars
plt.show()
```

```python
#MULTI BARGRAPHS
x=['python','c','java','cpp']
y=[50,47,84,16]
#can use same parameters as bar graphs
z=[20,30,40,50]
p=np.arange(len(x))#[1,2,3,4]
p1=[j+0.2 for j in p]#mark points just beside p(j+width) and draw the second graphs there
plt.bar(p,y,width=0.2,color='r',label='popularity1')
plt.bar(p1,z,width=0.2,color='y',label='popularity2')
#we get the numbers used in p as the x values
plt.xticks(p+0.2/2,x,rotation=10)
#set position of org values on x axis,give the names to be used in plce of p values(array numbers)
plt.legend(loc='upper left')#loc is position of labels
plt.show()
```

```python
#HORIZONTAL BAR GRAPH
#BAR GRAPH
x=['pyhton','c','java','c++']
y=[4,1,5,12]
c=['r','y','b','g']
plt.xlabel("language",fontsize=20)
plt.ylabel("populas",)
plt.title("lang and pop")
plt.barh(x,y,color=c,align="edge",edgecolor='r',linestyle=":",alpha=0.4,label="mygraph")
plt.legend()
plt.show()
```

```python
#SCATTER PLOT
day=[1,2,3,4,5,6,7]
no=[4,6,7,8,3,6,9]
no2=[1,2,3,4,5,7,8]
color=['r','g','b','y','r','g','b']
color1=[10,20,40,60,32,56,23]
sizes=[100,200,300,400,500,600,700]
# plt.scatter(day,no,color='r')
#if we use c=color{with codes} that is applied
#if we use c=colors[numbers] cmap will define the colors
plt.scatter(day,no,s=sizes,c=color1,alpha=0.7,marker="*",edgecolor="r",linewidth=1,cmap="viridis")
plt.scatter(day,no2,s=sizes,color="r",edgecolor="r",linewidth=1)
# plt.colorbar()
plt.title('scatter plot')
t=plt.colorbar()
t.set_label("color bar of my plot")
#to give label to colorbar
#graph parameters same as bar graph
plt.show()
```

```python
#PIE CHART
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.3,0.0,0.0,0.0]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",
        shadow=True,radius=1,labeldistance=1.2,startangle=90,
       textprops={"fontsize":10,"color":"c"},counterclock=False,
       wedgeprops={"linewidth":2,"edgecolor":'m'},center=(3,3),rotatelabels=False)
#"width":1 inside wedgeprops make it a hollow pie chart
plt.title("pie chart")
plt.legend(loc=2)
plt.show()
```

```python
#STEM PLOT
x=[1,2,3,4,5,6]
y=[2,3,4,5,6,2]
plt.stem(x,y,linefmt="-.",markerfmt="r*",bottom=2,basefmt="g",label="python",
        orientation="horizontal")
#use_line_collection=True
#linefmt=":"(dotted line),"-"(solid line),"--"(dashed line),"-."(dash dot line)
plt.legend()
plt.show()
```

```python
#STACK PLOT
x=[1,2,3,4,5,6]
area=[1,2,4,5,1,8]
area1=[5,2,7,9,1,6]
area2=[3,5,8,1,7,5]
l=["area","area1","area2"]
plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","y"],baseline="wiggle")
plt.title("python")
plt.legend()
plt.grid()
plt.show()
```

```python
#STEP PLOT
x=[1,3,4,6,7,8]
y=[2,6,8,2,8,2]
plt.step(x,y,color="r",marker="o",ms=10,mfc="green")
plt.grid()
plt.show()
```

```python
#FILL BETWEEN PLOT
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="r")
# plt.fill_between(x,y)
# plt.fill_between(x=[2,4],y1=3,y2=5,color="green")
plt.fill_between(x,y,color="yellow",where=(x>3)&(x<6),alpha=0.5)
plt.show()
```
